# Advent of Code 2022 - 14

URL: https://adventofcode.com/2022/day/14

We'll study ngn's [solution](https://codeberg.org/ngn/k/src/branch/master/aoc/22/14.k):

```ngnk
x:(|'.'" -> "\)'0:"i/14"
h:2+|/*+,/x /height
w:1000      /width
a:(&h*w){@[x;w/m+!1+(|/y)-m:&/y;:;1]}/,/2':'x /flat grid
d:0 /done? (fallen off the grid for part1)
:r:{$[d;0;a x;[d|:^a x;0];(a[x]:~d)+/o'x+w+0 -1 1]}500 /part1
r+{$[a x;0;(a[x]:1)+/o'0 -1 1+w+x]}500 /part2
```

The key to this problem is to picture the vertical slice of the cave, and phisically fill it with sand.
ngn assigns it to `a`, and later start filling it with sand until d==1

First, read the inputs, split by " -> ", convert the strings of numbers to actual numbers, and reverse the order.
Note that commas are evaluated as a "cons", so we effectively get the 2 numbers split

```ngnk
:x:(|'.'" -> "\)'10#0:"i/14"
```

`h` is the maximum height +2

```ngnk
:h:2+|/*+,/x /height
```

`w` is fixed to 1000

```ngnk
w:1000      /width
```

Here we want to create the vertial section of the cave: 

```ngnk
a:(&h*w){@[x;w/m+!1+(|/y)-m:&/y;:;1]}/,/2':'x /flat grid
```

We'll split the line by parts. 

First part is to iterate for each line, stencil of 2 elements, so we have point a to point b for each segment. 

```ngnk
:yy:10#,/2':'x
```

At this point, we draw the segments in a 1d array, having 1's where there is rock, and 0 where there is air. 

In each iteration:

- get the min of both x and y between the 2 points
- assign that minimum to `m`
- get the max of both x and y between the 2 points
- subtract one from the other, so you get a tuple of x y, where either x or y will be 0 
- Increment both numbers by one. That's very smart, you'll see
- call range with the tuple, that has a 1 in one and the diff +1 in the other, this will create a rank 2 matrix with zeros in one row, and numbers from 0 to the length of the segment.
- add the minimum point, so we have now the coordinates of the points to draw
- `w/`, encode with the base of `w`, meaning that for each different row, we increase `w` to the index of the array that contains the section.

```ngnk
y:((54 481;47 481))
:m:&/y
(|/y)-m:&/y
1+(|/y)-m:&/y
!1+(|/y)-m:&/y
m+!1+(|/y)-m:&/y
```

All this runs under a reduce function, passing the blank array representing the section of the cave in the beginning, and it gets filled up as we reduce through the entries of `x`.

```ngnk
:a:(&h*w){@[x;w/m+!1+(|/y)-m:&/y;:;1]}/,/2':'x /flat grid
```
